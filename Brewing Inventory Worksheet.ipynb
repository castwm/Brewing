```python
import pandas as pd
```

```python
# Build Lists to Convert to Dataframe
hops = []
malts = []
yeasts = []
adjuncts = []

labels = ['Ingredient', "Category", 'Amount', 'Unit', 'Store Type']

#Bolded Titles
def title(name):
    name = name
    bold ='\033[1m'
    print(bold + name)

    # Add different ingredient functions
def new_hop(name, amount, storage): 
    name = str(name)
    amount = amount
    storage = str(storage)
    hop = [name, 'Hop', amount, 'Ounces', storage]  #[Name, Category, Amount (oz), pellet or....]
    hops.append(hop)

def new_malt(name, amount, storage):
    name = str(name)
    amount = amount
    malt = [name, 'Malt', amount, 'Pounds', storage] #[Name, Category, Amount (oz), pellet or....]
    malts.append(malt)

def new_yeast(name, amount, storage):
    name = str(name)
    amount = amount
    yeast = [name, 'Yeast', amount, 'Packs', storage] #[Name, Category, Amount (oz), dry or wet]
    yeasts.append(yeast)

def new_adjunct(name, amount, measure, storage):
    name = str(name)
    amount = amount
    measure = str(measure)
    adj = [name, 'Adjunct', amount, measure, storage] #[Name, Category, Amount (oz), dry or wet]
    adjuncts.append(adj)
```

```python
# Build Inventory
new_hop('Pacific Jade', 0.5, 'Pellet')
new_hop('Motueka', 0.5, 'Pellet')
new_malt('Briess Pilsen M', 1.5, 'Grain')
new_malt('Goldpils Vienna Malt', 2, 'Grain')
new_malt('Light Roasted Barley', 0.75, 'Grain')
new_malt('Belgian Special', 0.75, 'Grain')
new_malt('Briess Carapils', 0.5, 'Grain')
new_adjunct('Corn Sugar', 3.5, 'Pounds', 'Crystals')

# Amber Ale Recipe Test
#new_malt('Light Roasted Barley', 0.75, 'Grain')
#new_malt('Belgian Special', 0.75, 'Grain')
#new_malt('Caramel 80L', 1, 'Grain')
#new_malt('Gold Malt Syrup', 6, 'Extract')
#new_hop('Willamette', 1, 'Pellet')
#new_yeast('Wyeast 1056', 1, 'Liquid')
#new_yeast("Munton's Ale Yeast", 1, 'Dry')

#print(hops)
#print(malts)
#print(inventory)
```

```python
#Build DataFrame of Inventory
hops_inventory = pd.DataFrame(data = hops, columns = labels)
malts_inventory = pd.DataFrame(data = malts, columns = labels)
yeasts_inventory = pd.DataFrame(data = yeasts, columns = labels)
adjuncts_inventory = pd.DataFrame(data = adjuncts, columns = labels)
inventory = pd.DataFrame(columns = labels)
inventory = inventory.append(hops_inventory)
inventory = inventory.append(malts_inventory)
inventory = inventory.append(yeasts_inventory)
inventory = inventory.append(adjuncts_inventory)

inventory.head(len(inventory))

#Set Index to ingredient name
inventory = inventory.set_index('Ingredient')
title('Inventory')
inventory.head(len(inventory))
```

```python
import pandas as pd
```

```python
hops = []
malts = []
yeasts = []
adjuncts = []

labels = ['Item', "Category", 'Amount', 'Store Type']

def new_hop(name, amount, storage):
    name = str(name)
    amount = amount
    storage = str(storage)
    hop = [name, 'Hop', amount, storage ]  #[Name, Category, Amount (oz), pellet or....]
    hops.append(hop)

def new_malt(name, amount, storage):
    name = str(name)
    amount = amount
    malt = [name, 'Malt', amount, storage] #[Name, Category, Amount (oz), pellet or....]
    malts.append(malt)

def new_yeast(name, amount, storage):
    name = str(name)
    amount = amount
    yeast = [name, 'Yeast', amount, storage] #[Name, Category, Amount (oz), dry or wet]
    yeasts.append(yeast)
```

```python
new_hop('New Zealand Pacific Jade', 0.5, 'Pellet')
new_hop('Motueka', 0.5, 'Pellet')
new_malt('Briess Pilsen M', 1.5, 'Grain')
new_malt('Goldpils Vienna Malt', 2, 'Grain')
new_malt('Light Roasted Barley', 0.75, 'Grain')
new_malt('Belgian Special', 0.75, 'Grain')
new_malt('Briess Carapils', 0.5, 'Grain')
print(hops)
#print(malts)
#print(inventory)
```

```python
hops_inventory = pd.DataFrame(data = hops, columns = labels)
malts_inventory = pd.DataFrame(data = malts, columns = labels)
yeasts_inventory = pd.DataFrame(data = yeasts, columns = labels)
adjuncts_inventory = pd.DataFrame(data = adjuncts, columns = labels)
inventory = pd.DataFrame(columns = labels)
inventory = inventory.append(hops_inventory)
inventory = inventory.append(malts_inventory)
inventory = inventory.append(yeasts_inventory)
inventory = inventory.append(adjuncts_inventory)
#inventory = inventory.iloc[2:]

#hops_inventory.head()
inventory.head(len(inventory))
```

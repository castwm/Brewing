```python
import pandas as pd
```

```python
# Build Lists to Convert to Dataframe
hops = []
malts = []
yeasts = []
adjuncts = []

labels = ['Ingredient', "Category", 'Amount', 'Unit', 'Store Type']

#Bolded Titles
def title(name):
    name = name
    bold ='\033[1m'
    print(bold + name)

    # Add different ingredient functions
def new_hop(name, amount, storage, where_to): # Name, amount, condition, list to add to
    name = str(name)
    amount = amount
    storage = str(storage)
    hop = [name, 'Hop', amount, 'Ounces', storage]  #[Name, Category, Amount (oz), pellet or....]
    where_to.append(hop)

def new_malt(name, amount, storage, where_to):
    name = str(name)
    amount = amount
    malt = [name, 'Malt', amount, 'Pounds', storage] #[Name, Category, Amount (oz), pellet or....]
    where_to.append(malt)

def new_yeast(name, amount, storage, where_to):
    name = str(name)
    amount = amount
    yeast = [name, 'Yeast', amount, 'Packs', storage] #[Name, Category, Amount (oz), dry or wet]
    where_to.append(yeast)

def new_adjunct(name, amount, measure, storage, where_to):
    name = str(name)
    amount = amount
    measure = str(measure)
    adj = [name, 'Adjunct', amount, measure, storage] #[Name, Category, Amount (oz), dry or wet]
    where_to.append(adj)
```

```python
#Build Hops Invetory
#Ensure Entries are Added Alphabetically
new_hop('Goldings', 0, 'Pellet', hops)

new_hop('Motueka', 0.5, 'Pellet', hops)

new_hop('Nelson Sauvin', 0, 'Pellet', hops)

new_hop('Pacific Jade', 0.5, 'Pellet', hops)

new_hop('Sorachi Ace', 0, 'Pellet', hops)

new_hop('Willamette', 0, 'Pellet', hops)

#print(hops)
```

```python
# Add Ajuncts to Inventory
new_adjunct('Corn Sugar', 3.5, 'Pounds', 'Crystals', adjuncts)

#print(adjuncts)
```

```python
#Build DataFrame of Inventory
hops_inventory = pd.DataFrame(data = hops, columns = labels)
malts_inventory = pd.DataFrame(data = malts, columns = labels)
yeasts_inventory = pd.DataFrame(data = yeasts, columns = labels)
adjuncts_inventory = pd.DataFrame(data = adjuncts, columns = labels)
inventory = pd.DataFrame(columns = labels)
inventory = inventory.append(hops_inventory)
inventory = inventory.append(malts_inventory)
inventory = inventory.append(yeasts_inventory)
inventory = inventory.append(adjuncts_inventory)

inventory = inventory.reset_index(drop = True)
#Set Index to ingredient name
#inventory = inventory.set_index('Ingredient')

working_inventory = inventory[inventory.Amount > 0]
working_inventory = working_inventory.reset_index(drop = True)

out_of = inventory[inventory.Amount <= 0]
out_of = out_of.reset_index(drop = True)

title('Inventory')
inventory.head(len(inventory))

#title('Working Inventory')
#working_inventory.head(len(working_inventory))

#title('Out Of')
#out_of.head(len(out_of))
```

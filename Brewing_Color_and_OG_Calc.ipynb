```python
from PIL import Image
import pandas as pd
pd.set_option('display.max_rows', None)

def title(name):
    name = name
    bold ='\033[1m'
    print(bold + name)
    
columns = ['Grain', 'Lovibond', 'Gravity lb/gallon', 'Ext. Eff. %']
           #Ext. Eff.% is extract efficiency

# DataFrame of All Grains, Lovibond and Gravity
all_grains = pd.read_csv(r"C:\Users\Will\Desktop\Brewing-main\Brewing-main\grain_values.csv",
                         names = columns)
all_grains.head(len(all_grains))
all_grain_to_
```

```python
#select grains for recipe
recipe_grains = pd.DataFrame(columns = columns)

def get_grain(grain, amount): #choose Grain and recipe to add to
    grain = str(grain)
    amount = float(amount)
    recipe = pd.DataFrame(columns = columns)
    index = all_grains.index[all_grains['Grain']== grain]
    row = all_grains.iloc[index][:]
    row['Amount'] = amount
    return row
# List of Grains
grain_list = ([['2-row pale malt',5],['maris otter',3.5],['flaked oats',1.5],
              ['flaked wheat', 1.5], ['cara red', 1.5]]) #Grain, amount
labels = ['Grain', 'Lovibond', 'Gravity lb/gallon', 'Ext. Eff. %', 'Amount']
recipe = pd.DataFrame(columns = labels)
#Loops to add grains to new dataframe
for x in grain_list:
    new_row = pd.DataFrame(get_grain(x[0], x[1]))
    recipe = recipe.append(new_row, ignore_index = True)
recipe.head()
```

```python
# Color Calculation

#MCU Calculation
mcus = []
def mcu(lovibond, amount):
    amount = amount
    lovibond = lovibond
    mcu = amount * lovibond / 5 # 5 is the number of gallons brewed
    mcus.append(mcu)

x = 0
while x < len(recipe):
    mcu(recipe.iloc[x]['Lovibond'], recipe.iloc[x]['Amount'])
    x+=1

# Get SRM
def srm(mcus):
    mcus = mcus
    srm = 0
    for x in mcus:
        srm+=x
    return srm

print(title('SRM Score: ' + str(srm(mcus))))

#Show SRM Beer Colors
im = Image.open(r"C:\Users\Will\Desktop\srm-chart-1035x1536.png")
im.show()
```
